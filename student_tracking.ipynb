```python
fs = []
ds = []
aws = []

def call_last_infos():
    try:
        with open("students_list.txt", "r") as file:
            for line in file:
                name, last_name, no, class_name = line.strip().split(" --- ")
                if class_name == "fs":
                    fs.append((name, last_name, no))
                elif class_name == "ds":
                    ds.append((name, last_name, no))
                elif class_name == "aws":
                    aws.append((name, last_name, no))
    except FileNotFoundError:
        # If the file is not found, create it and write an empty string to it.
        open("students_list.txt", "w").close()
    except Exception as e:
        print("There is a mistake:", e)

call_last_infos()


def add_record():
    """adds a new student enrollment"""
    name = input("Please enter a name..:").title().strip()
    last_name = input("Please enter a last name..:").title().strip()
    no = input("Please enter student number..:")
    while True:
        class_name = input("Please select a class name: (fs/ds/aws) ")
    
        if class_name == 'fs':
            fs.append((name, last_name, no))
            break
        elif class_name == 'ds':
            ds.append((name, last_name, no))
            break
        elif class_name == 'aws':
            aws.append((name, last_name, no))
            break
        else:
            print("Invalid input. Please enter a valid class name.")


    try:
        with open("students_list.txt", "a") as file:
            file.write(name + " --- " + last_name + " --- " + no + " --- " + class_name + "\n")
            print(name, last_name, "added.")
    except Exception as e:
        print("There is a mistake:", e)
    

def delete_record():
    """deletes a student enrollment"""
    no = input("Please enter the number of the student to be deleted..: ")
    try:
        with open("students_list.txt", "r") as file:
            lines = file.readlines()
        with open("students_list.txt", "w") as file:
            for line in lines:
                if no not in line.split(" --- "):
                    file.write(line)
            print(f"Student number {no} has been deleted.")
    except Exception as e:
        print("There is a mistake:", e)

def record_update():
    """updates a student record"""

    no = input("Please enter the number of the student..:")
    try:
        with open("students_list.txt", "r") as file:
            lines = file.readlines()
        with open("students_list.txt", "w") as file:
            for line in lines:
                if no not in line.split(" --- "):
                    file.write(line)
                else:
                    ad = input("New name: ").title()
                    last_name = input("New last name: ").title()
                    class_name = input("Please select a new class name (fs/ds/aws): ")
                    file.write(ad + " --- " + last_name + " --- " + no + " --- " + class_name + "\n")
                    print(f"Student number {no} has been updated.")
    except Exception as e:
        print("There is a mistake:", e)

def search_record():
    """searches for a student's record"""
    no = input("Please enter the number of the student..:")
    try:
        with open("students_list.txt", "r") as file:
            print("{:<15}{:<15}{:<10}{:<10}".format("name", "last name", "no", "class name"))
            print("-"*50)
            for line in file:
                name, last_name, no_, class_name = line.strip().split(" --- ")
                if no == no_:
                    print("{:<15}{:<15}{:<10}{:<10}".format(name, last_name, no_, class_name))
                    return
            print(f"Student number {no} not found.")
    except Exception as e:
        print("There is a mistake:", e)



while True:
    print("\n------ Welcome to the Student Tracking Program! ------")
    print("1. Add a new record")
    print("2. Delete a record")
    print("3. Update enrollment")
    print("4. Search a record")
    print("5. Show classes")
    print("6. Exit")
    choice = input("Your choice..: ")

    if choice == "1":
        add_record()

    elif choice == "2":
        delete_record()

    elif choice == "3":
        record_update()

    elif choice == "4":
        search_record()

    elif choice == "5":
        try:
            class_namelar = input("Which class? (fs/ds/aws) : ")
            if class_namelar == 'fs':
                if fs:
                    print(fs)
                else:
                    print("There are no students registered in class 'fs'.")
            elif class_namelar == 'ds':
                if ds:
                    print(ds)
                else:
                    print("There are no students registered in class 'ds'.")
            elif class_namelar == 'aws':
                if aws:
                    print(aws)
                else:
                    print("There are no students registered in class 'aws'.")
        except Exception as e:
            print("There is a mistake:", e)


    elif choice == "6":
        print("Exiting the program...")
        break

    else:
        print("Invalid input. Please enter the number of your choice..")

```
